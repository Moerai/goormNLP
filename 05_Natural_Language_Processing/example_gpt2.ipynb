```python
from transformers import GPT2LMHeadModel, GPT2TokenizerFast

tokenizer = GPT2TokenizerFast.from_pretrained('gpt2')
model = GPT2LMHeadModel.from_pretrained('gpt2')
```

```python
while True:
  nation = input("which country would you like to know about the capital? ")
  prompt = f"The capital of England is London\nThe capital of {nation} is"
  encodings = tokenizer(prompt, return_tensors='pt')
  generated = model.generate(**encodings)[0, encodings['input_ids'].shape[1]:]
  capital = tokenizer.decode(generated).split('\n', 1)[0].strip()
  print(capital)
```
